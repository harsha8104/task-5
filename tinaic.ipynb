```python
 1. Survival Count by Sex
sns.countplot(data=df, x='Sex', hue='Survived')


Observation:
- A significantly higher number of females survived compared to males.
- Most male passengers did not survive, showing a stark gender difference in survival rates.

🔹 2. Age Distribution
sns.histplot(data=df, x='Age', kde=True)


Observation:
- Most passengers were between 20 and 40 years old.
- The distribution is slightly right-skewed, with fewer elderly passengers.
- There’s a small peak in childhood ages, indicating a presence of younger passengers too.

🔹 3. Fare Distribution by Passenger Class
sns.boxplot(data=df, x='Pclass', y='Fare')


Observation:
- 1st class passengers paid considerably higher fares, with notable outliers.
- 3rd class fares had a tight distribution near the lower end, showing cost-effective tickets.
- Outliers in 1st class fares suggest a few extremely wealthy passengers.

🔹 4. Pairplot (Relationships Between Features)
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')


Observation:
- Survivors generally had higher fares and lower Pclass (1st class).
- A mix of young and older passengers survived, but class and fare appear stronger indicators than age alone.

🔹 5. Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')


Observation:
- Fare and Pclass show a strong negative correlation, indicating higher fares were tied to better class.
- Survival has weak-to-moderate positive correlation with Fare and negative correlation with Pclass.

✅ Summary of Findings:
- Gender played a significant role—women had notably higher survival rates.
- Passenger class and fare were key indicators of survival, with 1st class passengers having the best outcomes.
- Age distribution was centered on younger adults, but age itself was less predictive of survival.
- High outliers in fare and the correlation matrix support the idea that social status and ticket type influenced survival chances.
- No strong multicollinearity was observed, so features like age, fare, and class can be used together in modeling.


```
